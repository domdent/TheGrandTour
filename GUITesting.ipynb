```python
from pyqtgraph.Qt import QtCore, QtGui
import pyqtgraph.opengl as gl
import pyqtgraph as pg
import numpy as np
import sys
import time

class MainWindow(QtGui.QMainWindow):
    
    def __init__(self):
        super(MainWindow, self).__init__()
        self.setStyleSheet("QMainWindow{background: 'white'}")
        self.setGeometry(0,0,1500,1200)
        self.setWindowTitle("SVM")

        self.p = 0
        
        GrandTour = pg.PlotWidget()
        
        
        #add buttons
        Qbtn = QtGui.QPushButton("Quit", self) #just for testing
        Qbtn.clicked.connect(self.close_application)
        Qbtn.resize(Qbtn.minimumSizeHint())
        Qbtn.move(0,0)
        
        self.addPausebutton()
        
        #addmenu
        mainMenu = self.menuBar()
        #mainMenu.setNativeMenuBar(False) #puts menu bar in application
        fileMenu = mainMenu.addMenu('&File')
        playBackMenu = mainMenu.addMenu('&Playback')
        
        #actions to put on file menu
        extractAction = QtGui.QAction("&Fuit", self)
        extractAction.setShortcut("Ctrl+F")
        extractAction.triggered.connect(self.close_application)
        fileMenu.addAction(extractAction)
        
        extractAction = QtGui.QAction("&Leave App", self)
        extractAction.setShortcut("Ctrl+P")
        extractAction.triggered.connect(self.close_application)
        fileMenu.addAction(extractAction)
        
        #actions on playback menu
        extractAction = QtGui.QAction("&Change StepSize", self)
        extractAction.setShortcut("Ctrl+D")
        extractAction.triggered.connect(self.close_application)
        playBackMenu.addAction(extractAction)
        
        
        #add checkboxes todo make checkbox for each accuracy
        showSVM = QtGui.QCheckBox('SVM', self)
        showSVM.move(0,400)
        showSVM.stateChanged.connect(self.plotSVM)

        
        #add progressbar todo link to tour 
        self.progress = QtGui.QProgressBar(self)
        self.progress.setGeometry(0,1100,200,10)
        btn = QtGui.QPushButton("TestBTN", self) #just for testing
        btn.move(0,300)
        btn.clicked.connect(self.startTour)
        
        
        comboBox = QtGui.QComboBox(self)
        comboBox.addItem("1")
        comboBox.addItem("2")
        comboBox.addItem("3")
        comboBox.move(0,500)

        self.show()
    
        
    def startTour(self):
        self.completed = 0
        while self.completed < 100:
            self.completed += 0.001
            self.progress.setValue(self.completed) #for progress bar
            
    
    def plotSVM(self,state):
        if state == QtCore.Qt.Checked:
            print('Plotting SVM')
        else: print('Do not plot svm')
            
    def addPausebutton(self):
        self.pText = "Pause"
        self.pBut = QtGui.QPushButton(self.pText, self)
        self.pBut.clicked.connect(self.pause)
        self.pBut.resize(self.pBut.minimumSizeHint())
        self.pBut.move(0,100)
        
    def pause(self): 
        if self.p == 0:
            self.p = 1
            print(self.p)
            self.pBut.setText("Play")
        elif self.p == 1:
            self.p = 0

            print(self.p)
            self.pBut.setText("Pause")
    
    def close_application(self):
        #ask if you really want to leave
        choice = QtGui.QMessageBox.question(self, '',"Are you sure you want to quit?", QtGui.QMessageBox.No | QtGui.QMessageBox.Yes)
        if choice == QtGui.QMessageBox.Yes:
            print("Goodbye")
            sys.exit()
        else: pass
        

        
        
        
        
        
        
        
        
        
        
def main():
    app = QtGui.QApplication(sys.argv)
    GUI = MainWindow()
    sys.exit(app.exec_())
    
    

    
if __name__ == '__main__':
    main()
    

#sys.exit(app.exec_())
```
