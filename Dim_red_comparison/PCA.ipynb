```python
import numpy as np
import sklearn as sk
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/dom/Documents/MPhys/TheGrandTour/Data/wineData.txt", delimiter="\t")

data = np.array(df);
data = np.delete(data, 0, 0)
data = data.astype(float)
data = np.swapaxes(data,0,1)

Y = data[13]
X = np.delete(data, 13, axis=0)
X = X.T
```

```python
data_colour = []
data_marker = []
for i in range(len(Y)):
    if Y[i] == 1:
        data_colour.append("r")
        data_marker.append("x")
    elif Y[i] == 2:
        data_colour.append("b")
        data_marker.append("o")
    elif Y[i] == 3:
        data_colour.append("g")
        data_marker.append("*")
```

```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X.T)

np.shape(X)
```

```python
num_comps = 10

pca = PCA(n_components=num_comps)
principalComponents = pca.fit_transform(X)
print(pca.components_)



#print(principalComponents, np.shape(principalComponents))
#print("----------")
#print(X, np.shape(X))

#print(np.shape(principalComponents))

principalComponents = principalComponents.T
for i in range(num_comps - 1):
    plt.scatter(principalComponents[i], principalComponents[i+1], color=data_colour)
    plt.show()

```
